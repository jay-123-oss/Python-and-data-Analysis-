```python
import matplotlib.pyplot as plt

x = [1, 1]       # X-axis values
y = [1, 10, ]   # Y-axis values

plt.b(x, y)         # Line graph बनाना
plt.xlabel("X-axis")   # X-axis का नाम
plt.ylabel("Y-axis")   # Y-axis का नाम
plt.title("Simple Graph")  # ग्राफ का title
plt.show()             # ग्राफ को display करना
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3.5, 4.5, 5]       # X-axis (उदाहरण: पढ़ाई के घंटे)
y = [10, 15, 25, 38, 50]  # Y-axis (उदाहरण: मिले हुए marks)

plt.scatter(x, y)         # Scatter plot draw करता है
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.title("Study Hours vs Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))

# 1st Plot (1 row, 2 columns, 1st position)
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], [2, 4, 6])
plt.title("Plot 1")

# 2nd Plot (same figure, 2nd position)
plt.subplot(2, 2, 2)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title("Plot 2")

# 2nd Plot (same figure, 2nd position)
plt.subplot(2, 2, 3)
plt.bar([1, 5, 3], [1, 4, 2])
plt.title("Plot 2")

# 2nd Plot (same figure, 2nd position)
plt.subplot(2, 2, 4)
plt.scate([1, 2, 3], [1, 8, 9])
plt.title("Plot 2")
plt.show()

```

```python
import matplotlib.pyplot as plt

# 2 rows × 2 columns graphs
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

# Top-left
axes[0, 0].plot([1, 2, 3], [2, 5, 3])
axes[0, 0].set_title("1")

# Top-right
axes[0, 1].scatter([1, 2, 3], [3, 5, 7])
axes[0, 1].set_title("2")

# Bottom-left
axes[1, 0].bar([1, 2, 3], [5, 7, 3])
axes[1, 0].set_title("3")

# Bottom-right
axes[1, 1].hist([1, 2, 2, 3, 3, 3, 4])
axes[1, 1].set_title("4")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
sales = [10, 20, 15, 30, 25]
profit = [5, 10, 7, 15, 14]

plt.figure(figsize=(7, 4))

# Line 1
plt.plot(x, sales, 'b-o', label="Sales")
# Line 2
plt.plot(x, profit, 'r--', label="Profit")

plt.title("Sales vs Profit")
plt.xlabel("Days")
plt.ylabel("Amount in $")
plt.grid(True)
plt.legend()

plt.savefig("sales_profit_report.png")  # ✅ Save Graph
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [30, 20, 25, 25]
labels = ["Python", "Java", "C++", "JavaScript"]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=False)
plt.title("Programming Language Usage")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

plt.plot(x, y, label="Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Using xlim and ylim")

plt.xlim(0, 6)   # X-axis को 0 से 6 तक limit करो
plt.ylim(5, 35)  # Y-axis को 5 से 35 तक limit करो

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use('bmh')  # ✅ Stylish graph theme activate

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

plt.plot(x, y, marker='o')
plt.title("Graph with ggplot Style")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn')   # Style set

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

plt.plot(x, y, 'o-', label="Data Points")
plt.title("xlim, ylim and Style Example")
plt.xlabel("X")
plt.ylabel("Y")

plt.xlim(0, 6)         # axis limit
plt.ylim(0, 40)

plt.grid(True)
plt.legend()
plt.show()
```
